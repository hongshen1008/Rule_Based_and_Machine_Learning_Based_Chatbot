```python
# Greeting function

greet_in = ('hey', 'sup', "what's up", 'wassup', 'hi', 'hello', 'good day', 'bonjour', 'hola', 'greetings', 'yo')
greet_out = ['Hey', 'Hello', 'Hi there', 'Hi', 'Greetings', '*nods*']
def greeting(query):
    for word in query.split():
        if word.lower() in greet_in:
            return random.choice(greet_out) # return random greeting in greet_out
```

```python
def Naming(query):
    name = ''
    query = query.lower()
    if('my name is' in query):
        for i in query.split():
             if(i!='my' and i!= 'name' and i!='is'):
                name += ' ' + i.capitalize()
    elif('call me' in query):
        for i in query.split():
             if(i!='call' and i!= 'me'):
                name += ' ' + i.capitalize()
    elif('name is' in query):
        for i in query.split():
            if(i!= 'name' and i!='is'):
                name += ' ' + i.capitalize()
    elif('change my name to' in query):
        for i in query.split():
            if(i!= 'change' and i!='my' and i!= 'name' and i!='to'):
                name += ' ' + i.capitalize()
    elif('change my name to' in query):
        for j in query.split():
            if(i!= 'change' and i!= 'name' and i!='to'):
                name += ' ' + i.capitalize()
    elif('I am' in query or 'i am' in query):
        for i in query.split():
             if(i!='I' and i!= 'i' and i!= 'am'):
                name += ' ' + i.capitalize()
    else:
        for i in query.split():
            name += ' ' + i.capitalize()
    
    return name
```

```python
# Small talk function

Responses = {
    'how are you': 'I am fine. Thank you for asking',
    'how are you doing': 'I am fine. Thank you for asking',
    'how is it going': 'It is going great. Thank you for asking',
    'you alright': 'I am good. Thank you for asking',
    'good morning': 'Good Morning',
    'good afternoon': 'Good Afternoon',
    'good evening': 'Good Evening',
    'good day': 'Good day to you too',
    'whats up': "What's up",
    'sup': "What's up",
    'thanks': 'Dont mention it. You are welcome',
    'thankyou': 'Dont mention it. You are welcome',
    'thank you': 'Dont mention it. You are welcome',
    'my name': 'Your name is',
    'what is my name': 'Your name is',
    'what can you do': 'I can have a small talk, help to take order and answer some questions regarding the beverages!',
    'help': 'I can have a small talk, help to take order and answer some questions regarding the beverages!'
}
small_talk = Responses.values()
small_talk = [str (item) for item in small_talk]

# Stemming, vectorise and cosine similarity
def Smalltalk_Cosim(doc, query):
    query = [query]
    p_stemmer = PorterStemmer()
    bow = CountVectorizer()
    stemmed_doc = [p_stemmer.stem(w) for w in doc] #dictionary to list will only contain keys
    stemmed_query = [p_stemmer.stem(w) for w in query]
    bow_doc = bow.fit_transform(stemmed_doc)
    bow_query = bow.transform(stemmed_query)
    cosineSimilarities = cosine_similarity(bow_doc,bow_query).flatten()
    index = cosineSimilarities.argsort()[:-2:-1]
    if (cosineSimilarities[index] > 0.75):
        ans = [small_talk[i] for i in index[:1]]
        return ans[0]
```

```python
# Stemming and Vectorise
def stem_vectorise(doc, query):
    query = [query]
    p_stemmer = PorterStemmer()
    bow = CountVectorizer(stop_words=stopwords.words('english'))
    stemmed_doc = [p_stemmer.stem(w) for w in doc]
    stemmed_query = [p_stemmer.stem(w) for w in query]
    bow_doc = bow.fit_transform(stemmed_doc)
    bow_query = bow.transform(stemmed_query)
    return bow_doc, bow_query
```

```python
# Cosine Similarity
def cos_sim(a, b):
    cosineSimilarities = cosine_similarity(a, b).flatten()
    index = cosineSimilarities.argsort()[::-1]
    sim = cosineSimilarities.tolist()
    idx = index.tolist()
    sorted_sim = sorted(sim, reverse = True)

    # put index and similarity into dictionary
    idx_sim = {}
    for i, j in zip(idx, sorted_sim):
        idx_sim[i] = j

    idxSim_copy = idx_sim.copy() # make a copy of dictionary

    for x, y in idxSim_copy.items(): #delete item with similarity equal to 0
        if(y == 0.0):
            del idx_sim[x]
            
    idx = list(idx_sim.keys())
    sim = list(idx_sim.values())

    return sim, idx
```

```python
# Question Answering function

def questionAns(query):
    import csv
    # Read question-answer pair in csv file
    with open('QuestionsAnswering.csv', mode='r', encoding='utf-8-sig') as f:
        reader = csv.reader(f)
        corpus = {row[0]:row[1] for row in reader}

    #Stemming and Cosine Similarity
    a, b = stem_vectorise(corpus, query)
    sim, idx = cos_sim(a, b)
    ans = []

    idx_sim =  dict(zip(idx, sim))
    if all(item < 0.4 for item in sim):
        print("I am sorry, I cannot help you with this one. This might because your query does not appear in our menu. Try to key in 'menu' to see the drinks we offer. Cheers :)")
    else:
        for i, j in idx_sim.items():
            if (j > 0.45):
                ans += [list(corpus.values())[i]]
        if ans != None:
            print('\nGroovy: ')
            if(type(ans) == list):
                for i in ans:
                    print(i)
            else:
                print(ans)
```

```python
# Detect beverage from user query
def detectBeverage(query, menuPrice):
    from fuzzywuzzy import fuzz
    
    matching_items = []
    query = query.lower().split()
    for word in query:
        for item in menuPrice.keys():
            match_score = fuzz.token_set_ratio(word, item.lower()) # Calculate fuzzy match score between user query and coffee item
            if match_score > 80:
                if item not in matching_items:
                    matching_items.append(item.title())
    return matching_items

# detect number in user query
def detectNumber(query):
    num_dict = {"one": 1, "two": 2, "three": 3, "four": 4, "five": 5, "six":6, "seven":7, "eight":8, "nine":9, "ten":10} # number dictionary
    
    quantity = []
    for word in query.split():
        if word.isdigit():
            quantity += [int(word)]
        else:
            for w in num_dict:
                if w == word:
                    quantity += [num_dict[word]]
    return quantity

# transactional function
def transaction(query):

    menuPrice = {"Espresso": 1.50, "Americano": 1.50, "Cappuccino": 1.80, "Latte": 1.80, "Macchiato": 1.80, "Flat White": 2.10, "Mocha": 2.10, "Black Tea": 1.50}
    
    switch = True    
    while switch == True:
        # Detect quantity in user query
        temp_query = query
        quantity = detectNumber(temp_query)        
        
        # Detect beverage
        beverages = detectBeverage(query, menuPrice)

        if len(beverages) == 0:
            print('\nSorry, your order does not appear in our menu.')
            return True
        
        # Assume 1 for all beverage if no number detected
        if(len(quantity) == 0):
            for i in beverages:
                quantity += [1]   
    
        if len(quantity)!=0 and len(quantity) == len(beverages):
            # Find price of each beverage
            price = []
            for i, j in menuPrice.items():
                for k in beverages:
                    if k == i:
                        price += [j]

            #Calculate total price
            totalPrice = sum([p*q for p, q in zip(price, quantity)])

            print("\n{:<10} {:<15} {:<10}".format('Quantity', 'Beverage', 'Price'))
            print("----------------------------------\n")
            for i in range(len(price)):
                print("{:<10} {:<15} £{:<10.2f}".format(quantity[i], beverages[i], price[i]))
            print('\nTotal price is: £', "{:.2f}".format(totalPrice))
            
            flag = True
            while(flag == True):
                print('\nGroovy:\nGreat! Please confirm your item, quantity and total price (yes/no). If you want to exit transaction, type quit. :')
                confirmation = input().lower()
                if('yes' in confirmation):
                    flag = False
                    print("\nThank you! Your order has been confirmed.")
                    return False
                elif ('quit' in confirmation):
                    flag = False
                    return False
                elif('no' in confirmation):
                    reorder = input('\nWould you like to reorder (yes/no)?')
                    if ('yes' in reorder):
                        flag = False
                        return True
                    else:
                        flag = False
                        return False
                else:
                    print('\nGroovy:\nSorry I do not understand. Please type in "yes" or "no" to confirm your order.')
        else:
            print("Groovy:\nPlease specify the quantity of each item (e.g. 2 mocha and 1 americano). If you want to exit transaction, type quit.")
            query = input("\nUser:\n").lower()
            if "quit" in query:
                return False
            else:
                temp_query = query
```

```python
def Feedback(query):
    positive = ["good", "great", "not bad", "quite good", "its good", "it is good", "it's good", "perfect", "fantastic", "excellent", "informative and easy to use", "well done", "that was very helpful"]
    negative = ["bad", "it is bad", "its bad", "not good", "it is not good", "it can be improved further", "there are bugs", "there is a bug", "it is very bad", "terrible", "worst", "horrible", "not perform well"]
    
    if query in positive:
        print('\nGroovy: Thank you for your feedback, I am glad to hear that!')
    elif query in negative:
        print('\nGroovy: I am sorry to hear that, I will feedback to my company.')
    else:
        print('\nGroovy: Thank you for your feedback, we will improve our system in the future!')
```

```python
def chat():
    flag=True
    while(flag==True):
        print("Groovy:\nHello, welcome to Chin's Coffee House! What is your name?")
        n = input("\nUser:\n").translate(str.maketrans('', '', string.punctuation))
        name = Naming(n)
        newName = ''
        flag1=True
        while(flag1==True):
            print('\nGroovy:\nHi'+ (newName if len(newName)!=0 else name) +". Welcome to Chin's Coffee House. How can I help you? If you want to exit, type quit.")
            query = input("\nUser:\n")            
            temp = query
            query = query.lower()
            query = query.translate(str.maketrans('', '', string.punctuation))
            if(query=='quit'):
                flag1=False
                print("\nGroovy:\nThank you for visiting Chin's Coffee House. What do you think of my service?")
                query = input("\nUser:\n")
                Feedback(query)
                print("Bye, take care"+ (newName if len(newName)!=0 else name))
            elif('i am' in query or 'my name is' in query or 'call me' in query or 'name is' in query or 'change my name to' in query or 'change name to' in query):
                newName = Naming(temp)
            else:
                if(greeting(query)!=None):
                    print('\nGroovy: '+greeting(query) + (newName if len(newName)!=0 else name))
                elif(Smalltalk_Cosim(Responses, query)!=None):
                    x = Smalltalk_Cosim(Responses, query)
                    if('Your name is' in x and (len(name) == 0 and len(newName)==0)):
                        print('\nGroovy:\nYou have not enter a name. Plese enter a name to be called.')
                    else:
                        print('\nGroovy:\n'+x+(newName if len(newName)!=0 else name))
                elif('menu' in query or 'order' in query):
                    with open('Beverage_Menu.txt', 'r', encoding='utf-8') as file:
                        for line in file:
                            print(line)
                    flag2=True
                    while(flag2==True):
                        query = input('\nGroovy:\nWhat would you like to order? If you want to exit, type quit. :').lower()
                        query = query.translate(str.maketrans('', '', string.punctuation))
                        if query=='quit':
                            flag2=False
                        elif 'what' in query or 'what is' in query:
                            questionAns(query)
                        else:
                            flag2 = transaction(query)
                else:
                    questionAns(query)
        flag=False
```

## Run rule-based chatbot
